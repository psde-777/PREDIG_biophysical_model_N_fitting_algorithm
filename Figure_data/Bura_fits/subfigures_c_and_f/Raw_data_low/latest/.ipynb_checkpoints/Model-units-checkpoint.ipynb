# Calculating the volume surrounding a microfibril

The calculations done in this document are for the enzymatic hydrolysis samples used by Bura et al 

(R. Bura, R. Chandra, and J. Saddler, Influence of xylan on the enzymatic hydrolysis of steam-
pretreated corn stover and hybrid poplar, Biotechnology Progress, vol. 25, no. 2, pp. 315-322, 2009.)

The solution volume used was $V_\textrm{solution} = 50$mL.
The substrate was diluted at a weight-to volume consistency of 8%. The total substrate mass was therefore $m_\textrm{substrate} = 4$g.

The volume surrounding a microfibril is obtained by the formula $$ V_\textrm{surrounding} = \frac{ V_\textrm{solution} - V_\textrm{fibrils}}{N_\textrm{fibrils}},$$

where $V_\textrm{fibrils}$ and $N_\textrm{fibrils}$ are calculated as follows.

### Microfibril volume
The diameter of a glucose molecule is roughly $d_\textrm{glc} = 1$nm, and the distance between monomer bonds was assumed to be equal to this. Furthermore, a paper by Thomas et al (Structure of Cellulose Microfibrils in Primary Cell Walls
from Collenchyma) specifies the side length of a rectangular cellulose bundle of 24 polymers as roughly 3 nm. In order to include the outer layer of hemicellulose and lignin, this was increased to $d_\textrm{fibril} = 5$nm. From this, with the bond number $N_\textrm{bonds}$(here set to 100) the microfibril volume follows as $$ V_\textrm{fibril} = d_\textrm{fibril}^2 \cdot N_\textrm{bonds} \cdot d_\textrm{glc} $$ 


### Microfibril mass
The molar masses of the monomers are as follows (wikipedia):

$m_\textrm{glc} = 0.180156$kg/mol

$m_\textrm{xyl} = 0.15013$kg/mol

$m_\textrm{monolignol} = 0.180$kg/mol

The monolignol mass used is roughly in between the masses of the three main monolignols

The mass of the fibril is then calculated as $$ m_\textrm{fibril} = m_\textrm{glc} \cdot N_\textrm{glc} + m_\textrm{xyl} \cdot N_\textrm{xyl} + m_\textrm{monolignol} \cdot N_\textrm{monolignol} $$, where the monomer numbers are calculated from the percentages of cellulose,hemicellulose and lignin (see below). The percentages used here are for medium pre-treatment.

# Calculating the number of enzymes surrounding a microfibril

The concentration of each enzyme is assumed to be $c_\textrm{enzymes} = 1\mu$M $= 10^{-3}\frac{\textrm{mol}}{\textrm{m}^3}$. Multiplying this concentration with $V_\textrm{solution}$ yields the number of enzymes in the system. Below, some examples were calculated and plotted.

# Kinetic parameters

## Enzymes

Each of the four parameters has a concentration $[E_\textrm{n}]$ and a $k_\textrm{n}$ parameter associated to it. The $k_\textrm{n}$ denote the reaction rates of each enzyme, and their unit is $\frac{1}{\textrm{time}}$.
In the Gillespie-algorithm scheme, the $k_\textrm{n}$ are used to calculate the probality of a reaction of type n to happen inside a timestep of size $\delta t$:

$$ p_\textrm{n} = N_\textrm{substrate} \cdot N_\textrm{n} \cdot k_\textrm{n} \cdot \delta t $$

The propensity of a single reaction follows as 

$$ prop_\textrm{n} =  N_\textrm{n} \cdot k_\textrm{n} $$,

where summing over all $N_\textrm{n} = N_\textrm{substrate}$ reactions of type n yields $$ \sum_{i=1}^{N_\textrm{substrate}} = N_\textrm{substrate} \cdot N_\textrm{n} \cdot k_\textrm{n} = \frac{p_\textrm{n}}{\delta t} $$ 


This stochastic description translates directly to the deterministic scheme of the mass-action law, where the rate of change in reaction product is calculated as

$$ \dot{[P]} = \tilde{k_\textrm{n}} \cdot  [E_\textrm{n}] \cdot  [S], $$

with the enzyme concentration $[E_\textrm{n}]$, the substrate concentration $[S]$, and the reaction rate constant $\tilde{k_\textrm{n}}$, which for a reaction of type

$$ A + B \rightarrow \textrm{Products}$$ 

has units of $\frac{\textrm{volume}}{\textrm{time}}$. $k_\textrm{n}$ and $\tilde{k_\textrm{n}}$ are related by the simple relationship 

$$ \tilde{k_\textrm{n}} = V \cdot k_\textrm{n} $$


Since we use enzyme concentrations as input parameters, we need to define a reaction volume in order to arrive at the actual number of enzymes (or use a unit volume).

## Calculate mass-action product change and resulting product over time from substrate-over-time data

```python
K_react = 0.5
c_enzyme = 1


product_change_calc = np.full(substrate_var_data.shape,0.)
product_change_calc[:,0] = substrate_var_data[:,0]
product_change_calc[:,1]= K_react*c_enzyme*substrate_var_data[:,1]

product_integral = np.full(product_change_calc.shape,0.)
product_integral[:,0] = product_change_calc[:,0]
product_integral[0,1] = 0
for i in range(1,product_integral[:,1].size):
    dt = product_change_calc[i,0]-product_change_calc[i-1,0]
#    if dt < 0:
#        print(dt)
    product_integral[i,1] = product_integral[i-1,1] + dt*product_change_calc[i,1]
```

## Calculate numerical time-derivative of simulated cellobiose release data

```python
differentiated_prod_data = np.full(product_var_data.shape,0.)
differentiated_prod_data[:,0] = product_var_data[:,0]
#differentiated_prod_data[:,0] = np.linspace(0,product_var_data[-1,0],product_var_data[:,0].size)


dt = differentiated_prod_data[1,0] - differentiated_prod_data[0,0]
differentiated_prod_data[0,1] = (product_var_data[1,2]-product_var_data[0,2])/dt

differentiated_prod_data[1:-1,1] = (product_var_data[2:,2]-product_var_data[:-2,2])/((product_var_data[2:,0]-product_var_data[:-2,0]))

dt = differentiated_prod_data[-1,0] - differentiated_prod_data[-2,0]
differentiated_prod_data[-1,1] = (product_var_data[-1,2]-product_var_data[-2,2])/dt
```

## Lignin adhesive

There is only a single lignin adhesive reaction. As a result, its propensity depends on the concentration of lignin, that of all enzymes, and a k parameter. More precisely, the propensity reads as

$$ prop_\textrm{lign} = k_\textrm{lign} \cdot N_\textrm{lign} \cdot \sum_{n = 1}^4 N_\textrm{n} $$

$k_\textrm{lign}$ denotes the "rate constant of adhesion" of a single lignin monomer and also has units of $\frac{1}{\textrm{time}}$.

## Inhibition

The inhibiton by cellobiose and glucose is not implemented as one or more reactions, but instead as an effect that is active at every step. Currently it reduces the enzyme numbers depending on the respective inhibitor number and a weight parameter, by the simple relation:

$$ N_\textrm{blocked,EG} = w_\textrm{EG} \cdot N_\textrm{cellobiose} $$

$$ N_\textrm{blocked,CBH} = w_\textrm{CBH} \cdot N_\textrm{cellobiose}$$

$$ N_\textrm{blocked,BGL} = w_\textrm{BGL} \cdot N_\textrm{glucose} $$

The following plot shows the number of blocked EG- and CBH molecules for a range of inhibitor concentrations, with $w_\textrm{EG} = 0.025$, and $w_\textrm{CBH} = 0.05$.

```python
def N_blocked_EG_old(c_inhib,c_EG,c_CBH):
    affnity_EG = 0.1
    return (affnity_EG * c_inhib)
def N_blocked_CBH_old(c_inhib,c_EG,c_CBH):
    affnity_CBH = 1
    return (affnity_CBH * c_inhib)

c_EG_old  = 100
c_CBH_old = 100
N = 10000
N_blockeds_EG_old = []
N_blockeds_CBH_old = []
for i in range(N):
    c_inhib = i
#    c_EG = i
#    c_CBH = i
    EG_calc = c_EG_old - N_blocked_EG_old(c_inhib,c_EG_old,c_CBH_old)
    CBH_calc = c_CBH_old - N_blocked_CBH_old(c_inhib,c_EG_old,c_CBH_old)
    if EG_calc >= 0:
        N_blockeds_EG_old.append(EG_calc)   
    if CBH_calc >= 0:
        N_blockeds_CBH_old.append(CBH_calc)
#    print("inhib nr: " + str(c_inhib) + "; sum of blocked: " + str(N_blockeds_EG[i]+N_blockeds_CBH[i]) )
```

```python
with plotting_context("presentation"):
    plt.title("Fraction of blocked molecules for varying inhibitor concentration (old formulation)")
    plt.plot(range(len(N_blockeds_EG_old)),N_blockeds_EG_old)
    plt.plot(range(len(N_blockeds_CBH_old)),N_blockeds_CBH_old)
    plt.xlabel("c_inhibitor")
    plt.ylabel("N_blocked")
    plt.legend(("EG","CBH"))
    plt.show()
```

This linear way of increasing the blocking number is probably only valid for small inhibitor concentrations. It also does not take the enzyme concentration into account at all, and ignores the fact that one inhibitor works on two different enzymes. Furthermore, it does check whether the amount of blocked bonds exceeds the number of enzymes in the system (this case is considered in the simulations, where the enzyme concentration is set to 0 when it exceeds this).


A different formulation may be better: instead of linearly increasing the number of blocked enzymes depending only on the inhibitor number, a monod shape would probably make more sense:

$$ N_\textrm{blocked,EG} = w_\textrm{EG} \cdot \frac{N_\textrm{EG} \cdot N_\textrm{cellobiose}}{N_\textrm{EG} + N_\textrm{CBH} +N_\textrm{cellobiose}}  $$



$$ N_\textrm{blocked,CBH} = w_\textrm{CBH} \cdot \frac{N_\textrm{CBH} \cdot N_\textrm{cellobiose}}{N_\textrm{EG} + N_\textrm{CBH} +N_\textrm{cellobiose}}  $$


Here the number of blocked bonds initially rises quickly with the inhibitor number, after which the growth slows down gradually for larger inhibitor number. The weight factors now function as "binding rates" for the respective enzyme with the inhibitor. Their number is between 0 and 1 This formulation ensures that both enzyme concentrations are included in the calculation for each individual enzyme, and that the sum of blocked molecules does not exceed the number of inhibitors.

Product inhibition of cellulases studied with 14C-labeled cellulose substrates

```python
def N_blocked_EG(affnity_EG,c_inhib,c_EG,c_CBH):
    return (affnity_EG * c_EG * c_inhib/(c_EG + c_CBH + c_inhib))
def N_blocked_CBH(affnity_CBH,c_inhib,c_EG,c_CBH):
    return (affnity_CBH * c_CBH*c_inhib/(c_EG + c_CBH + c_inhib))
def N_blocked_BGL(affnity_BGL, c_inhib,c_BGL,c_placeholder=0):
    return(affnity_BGL * c_BGL*c_inhib/(c_BGL+c_inhib))


c_EG  = 1e-6
c_CBH = 1e-6
c_BGL = 1e-6*6.022*1e23

affnity_EG = 1
affnity_CBH = 1
affnity_BGL = 1

N = 1000
N_blockeds_EG = []
N_blockeds_EG_no_coupling = []
N_blockeds_CBH = []
N_blockeds_CBH_no_coupling = []
N_blockeds_BGL = []
for i in range(N):
    c_inhib = i
#    c_EG = i
#    c_CBH = i
    N_blockeds_EG.append(c_EG-N_blocked_EG(affnity_EG,c_inhib,c_EG,c_CBH))
    N_blockeds_EG_no_coupling.append(c_EG-N_blocked_BGL(affnity_EG,c_inhib,c_EG,c_CBH))
    N_blockeds_CBH.append(c_CBH-N_blocked_CBH(affnity_CBH,c_inhib,c_CBH,c_EG))
    N_blockeds_CBH_no_coupling.append(c_CBH-N_blocked_BGL(affnity_CBH,c_inhib,c_CBH,c_EG))
    N_blockeds_BGL.append(c_BGL-N_blocked_BGL(affnity_BGL,c_inhib,c_BGL,c_BGL))

#    print("inhib nr: " + str(c_inhib) + "; sum of blocked: " + str(N_blockeds_EG[i]+N_blockeds_CBH[i]) )
```

```python
outstuffs = np.full((N,2),0.)
outstuffs[:,0] = range(N)
#outstuffs[:,1] = N_blockeds_EG[:]
#outstuffs[:,2] = N_blockeds_EG_no_coupling[:]
#outstuffs[:,2] = N_blockeds_CBH[:]
#outstuffs[:,4] = N_blockeds_CBH_no_coupling[:]
outstuffs[:,1] = N_blockeds_BGL[:]
np.savetxt("BGL_affinity_2.txt",outstuffs,delimiter = "\t")
```
