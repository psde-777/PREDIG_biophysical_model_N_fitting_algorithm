```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
from qtbplot import plotting_context
```

```python
def logistic_decay(x,a,b,c):
    return (a/(1.+np.exp(-b*(x-c))))
    
```

```python
enzymes_50_data = np.loadtxt("lignin_vs_final_yield_enzymes_50.txt")
enzymes_100_data = np.loadtxt("lignin_vs_final_yield_enzymes_100.txt")
enzymes_150_data = np.loadtxt("lignin_vs_final_yield_enzymes_150.txt")

popt50, pcov50 = curve_fit(logistic_decay, enzymes_50_data[:,0], enzymes_50_data[:,1])
popt100, pcov100 = curve_fit(logistic_decay, enzymes_100_data[:,0], enzymes_100_data[:,1])
popt150, pcov150 = curve_fit(logistic_decay, enzymes_150_data[:,0], enzymes_150_data[:,1])
```

```python
with plotting_context("presentation"):
    plt.scatter(enzymes_100_data[:,0], enzymes_100_data[:,1])
    plt.plot(np.linspace(0,60,1000),logistic_decay(np.linspace(0,60,1000),popt100[0],popt100[1],popt100[2]))
    plt.ylim(-5,120)
    plt.legend(("fit", "data"))
    plt.show()
```
