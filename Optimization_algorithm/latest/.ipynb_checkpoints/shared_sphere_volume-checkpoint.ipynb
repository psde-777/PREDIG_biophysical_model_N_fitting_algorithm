# Calculating the volume shared by two overlapping spheres

This is used within the model to determine whether there is enough room around a bond to digest it. The formula found within the function v_shared(r1,r2,d), was obtained via the procedure outlined on https://mathworld.wolfram.com/Sphere-SphereIntersection.html.


```python
import numpy as np
import matplotlib.pyplot as plt
import qtbplot
from qtbplot import set_style, plotting_context
from qtbplot.colors import colorschemes
```

```python
def v_shared(r1,r2,d):
    if d >= r1+r2:
        return 0
    elif d < np.abs(r1-r2):
        if r1<r2:
            return 4*np.pi*r1*r1*r1/3
        else:
            return 4*np.pi*r2*r2*r2/3
    else:
        sol = np.pi*(r1+r2-d)*(r1+r2-d)*(d*d + 2*d*r2 - 3*r2*r2 + 2*d*r1 + 6*r2*r1 -3*r1*r1)
        return (  sol/(12*d))

# For reasoning behind this check out https://mathworld.wolfram.com/Sphere-SphereIntersection.html
    

def v_shared_longer(r1,r2,d):
    h1 = (r2-r1+d) * (r2+r1-d) / (2*d)
    h2 = (r1-r2+d) * (r1+r2-d) / (2*d)
    V1 = (1/3) * np.pi *h1*h1*(3*r1-h1)
    V2 = (1/3) * np.pi *h2*h2*(3*r2-h2)
    V_final = V1+V2
    return V_final
    
```

```python
r1 = 5
r2 = 3
d_start = 0
d_end = (r1+r2)*2
N_points = 1000
dists = np.linspace(d_start,d_end,N_points)
vol_1 = np.full(dists.size,4*np.pi*r1*r1*r1/3)
vol_2 = np.full(dists.size,4*np.pi*r2*r2*r2/3)
vols_quick = np.full(dists.size,0.)
for i in range(N_points):
    vols_quick[i] = v_shared(r1,r2,dists[i])
```

```python
with plotting_context("presentation", colorscheme="gb"):
    plt.title("Shared volume of two spheres")
    plt.plot(dists,vols_quick)
    plt.plot(dists,vol_1)
    plt.plot(dists,vol_2)
    plt.xlabel("Distance d (arbitrary units)")
    plt.ylabel("Shared volume (arbitrary units)")
    plt.legend(("quick","sphere 1", "sphere 2"))
    plt.show()
```
