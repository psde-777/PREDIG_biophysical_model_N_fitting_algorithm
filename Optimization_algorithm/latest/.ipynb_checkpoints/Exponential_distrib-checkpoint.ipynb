```python
import numpy as np
import matplotlib.pyplot as plt
import random
from qtbplot import set_style, plotting_context
from qtbplot.colors import colorschemes
```

```python
def trafo(r1):
    return(1/a0*(np.log(1/r1)))

N_randoms = 10000
a0 = 10
data = np.full((N_randoms,3),0.)



for i in range(N_randoms):
    r1 = random.random()
    data[i,0] = r1
    data[i,1] = trafo(r1);
    data[i,2] = 1/a0
#    print(random.random())
```

## This is the shape of the distribution we get if we take uniform random numbers and apply the transformation above

```python
with plotting_context("presentation"):
    plt.plot(data[:,0],data[:,1],".")
    plt.plot(data[:,0],data[:,2],".")
    plt.legend(("random distrib","1/a0"))
    plt.show()
```

```python
print(sum(data[:,1])/N_randoms)
```

```python
np.savetxt("expo_distrib.txt",data,delimiter = "\t")
```

## Now let us see, how often the respecive values on the y axis are actually picked!

```python
N_randoms = int(1e6)
N_intervals = int(5e3)
pick_nums = np.full((N_intervals,2),0.)
pick_nums[:,0] = np.linspace(0,1,N_intervals)
dx = pick_nums[1,0] - pick_nums[0,0]
for i in range(N_randoms):
    r1 = random.random()
    transformed = trafo(r1)
    multiplier = N_intervals
    transformed*=multiplier
    transformed = np.round(transformed)
#    print(int(N_intervals*transformed/multiplier))
    pick_nums[int(transformed),1]+=1
#    pick_nums[int(N_intervals*transformed/multiplier),1]+=1
#    for j in range(N_intervals):
#        if trafo(r1) >= pick_nums[j,0] and trafo(r1) < pick_nums[j,0]+dx:
#            pick_nums[j,1]+=1
pick_nums[:,1]/=N_randoms
```

```python
average_pick = np.average(pick_nums[:150,1])
```

```python
with plotting_context("presentation"):
    plt.plot(pick_nums[:,0],pick_nums[:,1],".")
    plt.plot(pick_nums[:,0],np.full(pick_nums[:,0].shape,1/a0))
    plt.plot(pick_nums[:,0],np.full(pick_nums[:,0].shape,average_pick))
    plt.legend(("random distrib","1/a0"))
    plt.show()
```

```python
import numpy as np
from collections import Counter
from bisect import bisect
from copy import deepcopy
N_randoms = int(1e6)
a0 = 1
values = np.random.uniform(0, 1, N_randoms)
values_save = deepcopy(values)
values = np.log(1/values)
values /= a0
average = np.average(values)


#breakpoints = np.linspace(0,1,int(1e3))
#breakpoints = (0.25, 0.5, 0.75, 1)
#keys = ["0 <= x < 0.25", "0.25 <= x < 0.5", "0.5 <= x < 0.75", "0.75 <= x <= 1"]
#keys = breakpoints
#data = dict(Counter(keys[bisect(breakpoints, i)] for i in values) )
```

```python
outdata = np.full((values.size,2),0.)
outdata[:,0] = values_save
outdata[:,1] = values
np.savetxt("expo_distrib_randoms.txt",outdata,delimiter="\t")
```
